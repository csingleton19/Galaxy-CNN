```python
def replace_hashtag(x):
    if '?' in x:
        return 'Uncertain'
    return x

galaxy_df['gz2class'] = galaxy_df['gz2class'].apply(replace_hashtag)


# galaxy_df = galaxy_df.applymap(lambda x: 'Uncertain' if isinstance(x, str) and x.startswith('U' or 'u') else x)
galaxy_df = galaxy_df.applymap(lambda x: 'Spiral' if isinstance(x, str) and x.startswith('S' or 's') else x)
galaxy_df = galaxy_df.applymap(lambda x: 'Elliptical' if isinstance(x, str) and x.startswith('E' or 'e') else x)
galaxy_df = galaxy_df.applymap(lambda x: 'Spiral' if isinstance(x, str) and x.startswith('A' or 'a') else x)
```
