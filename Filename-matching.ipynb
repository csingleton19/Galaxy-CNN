```python
import pandas as pd
import numpy as np
```

```python
galaxy_df = pd.read_csv('/home/cyrus/Documents/galaxy_df_cleaned.csv')
filename_mapping = pd.read_csv('/home/cyrus/Downloads/gz2_filename_mapping.csv')
```

```python
galaxy_df = galaxy_df.drop('Unnamed: 0', axis = 1)
galaxy_df
```

```python
filename_mapping
```

```python
galaxy_df = galaxy_df.merge(filename_mapping, left_on = 'dr7objid', right_on = 'objid', how = 'left')
galaxy_df.reset_index(drop = True)
```

```python
order = ['specobjid', 'dr8objid', 'dr7objid', 'rastring', 'decstring', 'gz2class', 'asset_id', 'sample_x', 'sample_y']
galaxy_df = galaxy_df[order]
galaxy_df
```
